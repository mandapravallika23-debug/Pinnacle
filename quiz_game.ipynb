```python
%%writefile QuizGame.java

import java.util.Scanner;

public class QuizGame {
    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);
        int score = 0;

        System.out.println("===== JAVA QUIZ GAME =====");

        System.out.println("\n1. What is the capital of India?");
        System.out.println("a) Mumbai");
        System.out.println("b) New Delhi");
        System.out.println("c) Chennai");
        System.out.print("Answer: ");
        String ans1 = sc.next();

        if(ans1.equalsIgnoreCase("b"))
            score++;

        System.out.println("\n2. Which language is used for Android development?");
        System.out.println("a) Java");
        System.out.println("b) HTML");
        System.out.println("c) CSS");
        System.out.print("Answer: ");
        String ans2 = sc.next();

        if(ans2.equalsIgnoreCase("a"))
            score++;

        System.out.println("\n3. How many days are there in a week?");
        System.out.println("a) 5");
        System.out.println("b) 6");
        System.out.println("c) 7");
        System.out.print("Answer: ");
        String ans3 = sc.next();

        if(ans3.equalsIgnoreCase("c"))
            score++;

        System.out.println("\nYour Score: " + score + "/3");

        sc.close();
    }
}
```
